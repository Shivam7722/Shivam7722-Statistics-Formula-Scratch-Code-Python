### Variance

```python
import pandas as pd

df = pd.DataFrame({'mileage':[19,19,20,20,20,21,21,21,
                              22,23,24,25,25,27,28,32,33,34,35]})
df
```

```python
def variance(a):
    
    #calculating mean
    sum_of_observation = 0
    for i in a['mileage']:
        sum_of_observation += i
    mean = sum_of_observation/len(a['mileage'])
    
    #calculating squared deviation
    squared_deviation = list()
    for i in a['mileage']:
        v = i-mean
        squared_deviation.append(v**2)
    
    # calculating sum of all squared deviations
    sum_of_squared_deviation = 0
    for i in squared_deviation:
        sum_of_squared_deviation += i
    
    # total number of observations
    n = len(a)
    
    #formula for variance
    pv = (1/n)*(sum_of_squared_deviation)
    sv = (1/(n-1))*(sum_of_squared_deviation)
    
    print(f'1. Population Variance mileage data is : {pv}\n2. Sample Variance mileage data is : {sv}')
        
        
variance(df)
```
